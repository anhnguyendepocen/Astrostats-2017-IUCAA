The appropriate likelihood is the binomial distribution
\begin{equation}
P(r | p, n, M) \,=\, \left(
   \begin{array}{c}
    n \\
    r
  \end{array} \right)  \, p^r(1-p)^{n-r}
\end{equation}
where $r \leq n$.

\begin{equation}
P(r | p, n, M) \,=\, \sum_j x_j
\end{equation}

```r
par(mfrow=c(1,1), mgp=c(2.0,0.8,0), mar=c(3.5,3.5,1,1), oma=0.1*c(1,1,1,1))
n <- 20
r <- 7
p <- seq(from=0,to=1,length.out=201)
plot(p, dbinom(x=r, size=n, prob=p), ylim=c(0, 0.2), xaxs="i", yaxs="i", 
  xlab="p", ylab=expression(paste(P^symbol("*"), "(p | r,n,M)")), type="l", 
  lwd=1.5)
```

The following R code implements the mean as just described and overplots it in figure \ref{fig:coin0} with the dashed line.

```r
pdense <- dbinom(x=r, size=n, prob=p) # unnormalized in p
p.mean <- sum(p*pdense)/sum(pdense)
cat(p.mean)
#abline(v=p.mean, lty=2)
```
